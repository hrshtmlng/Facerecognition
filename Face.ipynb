# Face Detection and Gender Recognition

This notebook demonstrates face detection using OpenCV and gender recognition using DeepFace.

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
img = cv2.imread('/content/OIP (1).jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

cv2_imshow(img)
```

```python
from deepface import DeepFace
img_path = "/content/OIP.jpg"
obj = DeepFace.analyze(img_path, actions=['gender'])
print(f"The detected gender is: {obj[0]['gender']}")
```
